```python

 #Fabric Lakehouse: read CSV from Files and create a managed Delta table under Tables
 
from notebookutils import mssparkutils

raw_path = "/lakehouse/default/Files/raw/sample_raw_sales.csv"
table_name = "bronze_sales"   # table name kept exactly as requested

# 1) Verify the source file exists
assert mssparkutils.fs.exists(raw_path), f"File not found: {raw_path}"

# 2) Read and preview the CSV
df = (
    spark.read.format("csv")
    .option("header", "true")
    .option("inferSchema", "true")
    .load(raw_path)
)

df.printSchema()
display(df.limit(10))

# 3) Write as a managed Delta table (appears in Lakehouse -> Tables)
(
    df.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable(table_name)
)

# 4) Validate
print(f"Wrote managed table: {table_name}")
spark.sql(f"SELECT COUNT(*) AS rows FROM {table_name}").show()
display(spark.table(table_name).limit(10))
```
